```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

font_name = fm.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
plt.rc('font', family=font_name)

mpl.rcParams["axes.unicode_minus"] = False
```

### layer 조정

+ case1 = CNN layer 2, FC layer 1 droupout 20% , 20%, 40% , batch_size = 256, epoch=10 
+ case2 = CNN layer 3, FC layer 1 dropout 20%, 20%, 20%, 40%, epoch=10, batch_size=256 
+ case3 = CNN layer 2 , FC layer 2 dropout 20%, 20%, 40%, 40%, epoch=10, bath_size=256
+ case4 = CNN layer 3, FC layer 2 dropout 20% 20% 20% 40% 40%, epoch=10, batch_size=256 

```python
x = np.arange(4)
case = ['case1', 'case2', 'case3', 'case4']
values = [0.96, 0.96, 0.97, 0.93]

plt.title("Layer 수에 따른 정확도 비교")
plt.plot(x, values, marker='o')
plt.xticks(x, case)
plt.xlabel('Case', fontsize=13)
plt.ylabel('Accuracy', fontsize=13)
plt.show()
```

### dropout 조정

```python
x = np.arange(5)
case = ['case1', 'case2', 'case3', 'case4', 'case5']
values = [0.96, 0.97, 0.96, 0.96, 0.97]

plt.title("Droupout 조정에 따른 정확도 비교")
plt.plot(x, values, marker='o')
plt.xticks(x, case)
plt.xlabel('Case', fontsize=13)
plt.ylabel('Accuracy', fontsize=13)
plt.show()
```
